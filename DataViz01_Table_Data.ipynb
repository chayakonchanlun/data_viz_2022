```python
import pandas as pd 
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv') #vjko
df
```

```python
df['Name'].unique() #unique เลือกค่าที่ไม่ซำ้กันเช่นซื่อ พันธุ์ของดอกไม้
```

## Parallel coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name')#parallel_coordinates คือการเอาค่าแต่ละคอลัมมาพลอตเทียบกับ โดยใช้ colom ในการแยกโดยใช้สี แสดงกรีบเลี้ยงกรีบดอก
```

```python
pd.plotting.parallel_coordinates(df, 'Name ', cols = ['SepalLength' , 'PetalLength'])
```

## Change color

```python
pd.plotting.parallel_coordinates(df,'Name',color=['r','y','k'])
```
