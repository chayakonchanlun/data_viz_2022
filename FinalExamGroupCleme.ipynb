##เลือกเฉพาะคอลัมน์ที่สนใจ

```python
alldb1 = alldb[['user','RowKey','date','ov','phone','site','time','urine','PartitionKey']]  #เลือกคอลัมน์ที่เราสนใจ
alldb1
```

##กำจัดค่า Missing

```python
alldb1.isnull().any()  #ขึ้น True แสดงว่ายังมีค่า Missimg 
```

```python
alldb2 = alldb1.dropna()  #จัดการ Missing ด้วยการ dropna
alldb2 
```

```python
removes = alldb1.shape[0] - alldb2.shape[0]   #เชคว่าจากการดรอปข้อมูลหายกี่เปอร์เซนต์ ซึ่งพบว่าหาย 3.427001373797927% ของข้อมูลทั้งหมด ยอมรับได้
print(f'size before drop = {alldb1.shape[0]}')
print(f'size aftre drop = {alldb2.shape[0]}')
print(f'we loss {100*(removes/alldb1.shape[0])}% of data')
```

```python
print(alldb1.shape)   #จำนวนแถวก่อนDropna
print(alldb2.shape)   #จำนวนแถวหลังDropna
```

```python
alldb2.isnull().any()   #เชค Missing พบว่าขึ้น False แสดงว่าไม่มี Missing แล้ว
```

##จัดการข้อมูลให้เป็นไปตามข้อกำหนดของอาจารย์

###User เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)



```python
alldb3 = alldb2[(alldb2['user'].astype(str).str.isdigit()) & (alldb2['user'].str.len() == 5)]  #เลือกข้อมูลที่เป็นตัวเลข 5 หลักเท่านั้น
alldb3
```

###RowKey เป็น String

```python
set(alldb3['RowKey']) #ดูค่าในคอลัมน์ RowKey
```

```python
alldb3['RowKey'] = alldb2['RowKey'].astype('string')  #เปลี่ยนชนิดข้อมูลโดยใช้ astype
alldb3
```

```python
print(alldb3.dtypes)  #เชคชนิดข้อมูล RowKey เปลี่ยนเป็น  string แล้ว
```

###date เป็น YYYY-MM-DD

```python
set(alldb3['date'])  #ดูข้อมูลในคอลัมน์ date
```

```python
alldb3['date'] = alldb3['date'].replace(':','-', regex=True)  #แทนที่ : ด้วย - เพื่อให้เป็นตามข้อกำหนด
alldb3
```

###phone เป็น String

```python
set(alldb3['phone']) #ดูค่าข้อมูลในคอลัมน์ phone
```

```python
alldb3['phone'] = alldb3['phone'].astype('string')  #เปลี่ยนชนิดข้อมูลโดยใช้ astype
alldb3
```

###site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)


```python
alldb4 = alldb3[(alldb3['site'].astype(str).str.isdigit()) & (alldb3['site'].str.len() == 5)]  #เลือกข้อมูลที่เป็นตัวเลข 5 หลักเท่านั้น
alldb4
```

###time เป็น hh-mm-ss

```python
set(alldb4['time'])  #ดูข้อมูลในคอลัมน์ time 
```

```python
alldb4['time'] = alldb4['time'].replace(':','-', regex=True)  #แทนค่า : ด้วย - เพื่อให้เป็นตามข้อกำหนด
alldb4
```

###urine มีค่า String

```python
alldb4['urine'] = alldb4['urine'].astype('string')  #เปลี่ยนชนิดของข้อมูลให้เป็น string ด้วย astype
alldb4
```

```python
print(alldb4.dtypes)  #เชคชนิดข้อมูลพบว่าเป็น string แล้ว
```

###ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)
โดยจะกำหนดให้แทนด้วยตัวเลขดังต่อไปนี้ 
* 0 = ไม่พบโรค
* 1 = พบโรคที่ระดับความรุนแรง 1
* 2 = พบโรคที่ระดับความรุนแรง 2
* 3 = พบโรคที่ระดับความรุนแรง 3
* 4 = พบโรคที่ระดับความรุนแรง 4
* ติดเชื้อ +2 = พบโรคที่ระดับความรุนแรง 4

```python
set(alldb4['ov'])  #ดูข้อมูลคอลัมน์ ov
```

```python
alldb4['ov'].replace({'0':'ไม่พบโรค',
'1':'พบโรคที่ระดับความรุนแรง 1',
'2':'พบโรคที่ระดับความรุนแรง 2',
'3':'พบโรคที่ระดับความรุนแรง 3',
'4':'พบโรคที่ระดับความรุนแรง 4',
'ติดเชื้อ +2 ':'พบโรคที่ระดับความรุนแรง 4',}, inplace=True)  #แทนค่าข้อมูลตามที่เรากำหนด
```

```python
set(alldb4['ov'])  #เชคข้อมูลคอลัมน์ ov ข้อมูลเป็นตามที่กำหนดแล้ว
```

###PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)
* เลือกมาเฉพาะข้อมูลที่เป็นตัวเลข 5 หลัก

```python
set(alldb4['PartitionKey'])  #ดูค่าข้อมูลคอลัมน์ PartitionKey
```

```python
alldb5 = alldb4[(alldb4['PartitionKey'].astype(str).str.isdigit()) & (alldb4['PartitionKey'].str.len() == 5)]  #เลือกข้อมูลที่เป็นตัวเลข 5 หลักเท่านั้น
alldb5
```
