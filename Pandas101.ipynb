## **HW (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)**

1.   คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ของข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ของผู้ป่วยในจังหวัดขอนแก่น
3. หาจำนวนผู้ป่วยที่เป็นคน "**ไม่ใช่คนไทย**"

````python
this_data = data_covid[['sex','age','province_of_onset']] # ข้อมูลที่ใช้้
````

````python
this_data
````

````python
men = this_data[this_data['sex']=='ชาย']
men['age']
````

````python
sum = 0 #เอาsum มารับค่าทั้งหมดที่ต้องเริ่มต้นจาก 0
N = 0 #ตัวที่ใช้หาร .ใช้0
for a in men['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}')
````

````python
men_KK = men[men['province_of_onset']=='ขอนแก่น']

sum = 0
N = 0
for a in men_KK['age']:
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย ในจังหวัดขอนแก่น {sum/N}')
````

````python
#ข้อ 3
foreigner =  data_covid[data_covid['nationality'].notnull()] #สร้างตัวแปร foreigner รับค่า data_covid คอลัมน์ nationality โดยไม่เอาค่า NaN
foreigner_quiz = foreigner[foreigner['nationality']!='Thailand'] #สร้างตัวแปร foreigner_quiz รับค่า foreigner คอลัมน์ nationality โดยไม่เอา Thailand
num = foreigner_quiz['No.'] #กำหนด num รับค่า foreigner_quiz ที่แสดงค่า No.
foreigner_quiz[['No.','nationality']] #แสดงคอลัมน์ No and nationality
````

````python
#ข้อ 3 หาจำนวนผู้ป่วยที่เป็นคน "ไม่ใช่คนไทย"
sum = 0 #กำหนดให้ sum=2
for a in num: #วนลูป a ทั้งหมดในตัวแปร num
  if a>0: #ถ้า a>0 ให้ทำต่อ
    sum +=1 #ให้ +1 ใน sum ทุกครั้งที่เจอ a>0
print(f'ผู้ป่วยที่ไม่ใช่คนไทยทั้งหมด {sum} คน')
````

# Quiz ตัดตารางออกมาเฉพาะปี 2021 announce_date ในปี 2021

Hint

วนลูปหา index ของปี 2021

ตัดตารางมาเฉพาะ ปี 2021

````python
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````

HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20

````python
TF=list()
for each_row in data_covid.iterrows(): # หลัง in จะต้องเป็น list ที่เราอยากจะเอามาวนรูป โดยใช้ . iterrows ในการเเปลงค่าจากตาราง panda ให้มาเป็นค่า list
    if each_row[1]['announce_date'].split('/')[1] <= '20' : #ให้วนรูปตั้งเเต่คอลัม announce_date เเละตั้งเเต่เเถวที่ 1
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF]
````
